#### The following code creates the plot of most used languages per year

```python
#Merging the individual dataframes of most used language per year and identifying the top eight for each year
merged=pd.concat([df_2011,df_2012,df_2013,df_2014,df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021],axis=1)
merged.columns=['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
final=merged.head(8).T

#Generate plot
ax = final.plot.bar(figsize=(10,10),rot=0,stacked=True)
# To show data labels inside the bars
labels = [f'{i:.0f}' for i in final.to_numpy().flatten(order='F')]

for i, patch in enumerate(ax.patches):
    x, y = patch.get_xy()
    x += patch.get_width() / 2
    y += patch.get_height() / 2
    ax.annotate(labels[i], (x, y), ha='center', va='center', c='white')

ax.legend(loc='best',bbox_to_anchor=(1.2, 1))
ax.set_xlabel("Year",fontdict={'fontsize':12})
ax.set_ylabel("Cumulative Value of Percentage Preference",fontdict={'fontsize':12})
```

#### The following code creates the plot of top three salary values that is common for the majority of the participants (Question 3)

```python
#Merging the individual dataframes of most desired language per year and identifying the top 20 for each year
merged=pd.concat([df_2015_cur_val,df_2016_cur_val,df_2017_cur_val,df_2018_cur_val,df_2019_cur_val,df_2020_cur_val,df_2021_cur_val],axis=1)
merged.columns=['2015', '2016', '2017', '2018', '2019', '2020', '2021']
# The salary valeus were created as an index, and we are converting them as a column to plot the value
merged['index1'] = merged.index 
 
#Generate plot
fig, axes = plt.subplots(3, 3, figsize=(8, 8), constrained_layout=True, sharey=True)
for col, ax in zip(merged.columns, axes.flat):
    #We do not want to plat the salary value column, we are only plotting the number of participants per year
    if col!='index1':
        df=merged[['index1',col]]
        df=df.dropna(subset=[col])
        #Finding the top 3 salary values
        df=df.sort_values(by=[col], ascending=False).head(3)
        df[col].plot.bar(ax=ax, rot=0)
        ax.set_title(col)
        ax.set_ylabel("Number of Respondents",fontdict={'fontsize':12})
```
